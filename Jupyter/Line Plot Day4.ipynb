### First Step to read data from the file 
**import libraries**
- import pandas
- import seaborn
- import matplotlib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
day4=pd.read_csv("C:/Users/Yasir Mehmood/Downloads/Python Programs/Chilla_data2_for_plots.csv")
day4
```

### Draw a Line plot and its functions first

```python
sns.lineplot(x="Research/Working experience (Float/Int) years", y="Light kitni der band hti hy? int", data=day4)
plt.show()
```

### Adding Titles

```python
sns.lineplot(x="Research/Working experience (Float/Int) years", y="Light kitni der band hti hy? int", data=day4)
plt.title("Graph between Working experience and Light")
plt.show()
```

### Set Limits

```python
sns.lineplot(x="Research/Working experience (Float/Int) years", y="Light kitni der band hti hy? int", data=day4)
plt.title("Graph between Working experience and Light")
plt.xlim(0)
plt.ylim(1)
plt.show()
```
