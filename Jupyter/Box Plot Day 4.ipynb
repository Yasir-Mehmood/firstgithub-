### Show Labels

```python
import matplotlib.pyplot as plt
p1=sns.boxplot(x="Gender", y="Age (years)-Float/Int", showmeans=True,
               meanprops={"marker":"*", "markersize":"12", "markeredgecolor":"yellow"},
               data=day4)
plt.xlabel("How many Male and Female", size=10),
plt.ylabel("How many years", size=10),
plt.title("Bar Plot between Gender and their age", size=14)
```

```python
import matplotlib.pyplot as plt
p1=sns.boxplot(x="Gender", y="Age (years)-Float/Int", showmeans=True,
               meanprops={"marker":"*", "markersize":"12", "markeredgecolor":"yellow"},
               data=day4)
plt.xlabel("How many Male and Female", size=10, weight='bold'),
plt.ylabel("How many years", size=10, weight='bold'),
plt.title("Bar Plot between Gender and their age", size=14, weight='bold')
```
