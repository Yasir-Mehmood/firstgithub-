### Adding Tittles

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower = sns.load_dataset ("iris")
flower

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Flowers Plot")
plt.show()
```

### Adding Limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
flower = sns.load_dataset ("iris")
flower

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Flowers Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None,rc=None)
# Load dataset
flower = sns.load_dataset ("iris")
flower

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Flowers Plot")
sns.set_style("darkgrid")
plt.show()
```

### size of figures

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None,rc=None)
# Load dataset
flower = sns.load_dataset ("iris")
#change figure
plt.figure(figsize=(8,6))

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Flowers Plot")

plt.show()
```
