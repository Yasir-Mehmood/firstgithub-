### First Step to read data from the file 
**import libraries**
- import pandas
- import seaborn
- import matplotlib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
day4=pd.read_csv("C:/Users/Yasir Mehmood/Downloads/Python Programs/Chilla_data2_for_plots.csv")
day4
```

### Bar Plot

```python
sns.barplot(x="What are you?", y="Light kitni der band hti hy? int", data=day4)
plt.show()
```

### Check Graph using different elements

```python
sns.barplot(x="What are you?", y="Age (years)-Float/Int", data=day4)
plt.show()
```

### Add third element hue

```python
sns.barplot(x="What are you?", y="Age (years)-Float/Int", hue="Where do you live?", data=day4)
plt.show()
```

```python
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4)
plt.show()
```

### Add Order Element

```python
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4, order=["Rural","Urban",
                                                                                                      "International"])
plt.show()
```

### Add Color

```python
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4, order=["Rural","Urban",
                                                                                                      "International"],color="red")
plt.show()
```

### Add confidence interval Element

```python
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4,
            order=["Rural","Urban", "International"],color="red", ci=None)
plt.show()
```

### Add Palette Element

```python
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4,
            order=["Rural","Urban", "International"],color="red", ci=None, palette="bright")
plt.show()
```

### Add Mean Element

```python
from numpy import mean
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4,
            order=["Rural","Urban", "International"],color="red", estimator=mean, palette="bright")
plt.show()
```

### Add Median Element

```python
from numpy import median
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4,
            order=["Rural","Urban", "International"],color="red", estimator=median, palette="bright")
plt.show()
```

### Add Saturation element

```python
from numpy import median
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", hue="What are you?", data=day4,
            order=["Rural","Urban", "International"],color="red", estimator=median, saturation=4, palette="bright")
plt.show()
```

```python
# Horizontal Plot
from numpy import median
sns.barplot(x="Age (years)-Float/Int", y="Where do you live?", hue="What are you?", data=day4, estimator=median, saturation=2)
plt.show()
```

```python
sns.barplot(x="Where do you live?", y="Age (years)-Float/Int", data=day4, linewidth=2.5, facecolor=(0.2,0.3,0.4,0.5),
           errcolor=".2", edgecolor=".2")
plt.show()
```
